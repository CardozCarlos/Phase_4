```python
# Instalación de librerías
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carga el archivo que contiene los datos de nivel de los tanques
df = pd.read_csv("test_dataset.csv")

# Aquí se convierte los datos de "DATETIME" a formato de fecha y hora
df['DATETIME'] = pd.to_datetime(df['DATETIME'])

# Con estos mensajes se muestran las estadísticas de los tanques L_T1 y L_T2
print("\nEstadísticas de L_T1 y L_T2:")
print(df[['L_T1', 'L_T2']].describe())

# Ahora se grafíca el comportamiento de los niveles
plt.figure(figsize=(14, 5))
plt.plot(df['DATETIME'], df['L_T1'], label='Nivel L_T1', color='blue')
plt.plot(df['DATETIME'], df['L_T2'], label='Nivel L_T2', color='orange')
plt.title("Evolución del nivel de los tanques L_T1 y L_T2")
plt.xlabel("Fecha y hora")
plt.ylabel("Nivel del tanque")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Se grafican los histogramas de distribución de ambos tanques
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['L_T1'], kde=True, color='skyblue')
plt.title("Distribución del nivel L_T1")
plt.xlabel("Nivel")

plt.subplot(1, 2, 2)
sns.histplot(df['L_T2'], kde=True, color='coral')
plt.title("Distribución del nivel L_T2")
plt.xlabel("Nivel")

plt.tight_layout()
plt.show()
```
